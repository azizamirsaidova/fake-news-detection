```python
import pandas as pd
import re
import nltk.data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import spacy
from spacy import displacy
import collections
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
from nltk.tokenize import sent_tokenize
```

**Reading in data in txt format:**

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
false_results = []
with open('false_results.txt') as f:
    lines = f.readlines()
    for line in lines:
      line = line.replace("[{'generated_text': '", '').replace("[{'generated_text': \"", '')
      line = line.replace("[", '').replace("'}],", '').replace("\"}],", '')
      line = line.replace("\\n", ' ').replace('\\', '')
      false_results.append(line)

true_results = []
with open('true_results.txt') as f:
    lines = f.readlines()
    for line in lines:
      line = line.replace("[{'generated_text': '", '').replace("[{'generated_text': \"", '')
      line = line.replace("[", '').replace("'}],", '').replace("\"}],", '')
      line = line.replace("\\n", ' ').replace('\\', '')
      true_results.append(line)
```

```python
false_results[:5]
```

**Calculating Part of Speech Frequencies:**

```python
total_tokens_false = 0
total_tokens_true = 0
total_pronouns_false = 0
total_pronouns_true = 0
total_propernouns_false = 0
total_propernouns_true = 0
pronouns_false = {}
pronouns_true = {}
propernouns_false = {}
propernouns_true = {}
for s in false_results:
  doc = nlp(s)
  for token in doc:
    total_tokens_false+=1
    if token.pos_=='PRON':
      total_pronouns_false+=1
      if token.text.lower() in pronouns_false:
        pronouns_false[token.text.lower()]+=1
      else:
        pronouns_false[token.text.lower()]=1
    elif token.pos_=='PROPN':
      total_propernouns_false+=1
      if token.lemma_ in propernouns_false:
        propernouns_false[token.lemma_]+=1
      else:
        propernouns_false[token.lemma_]=1

for s in true_results:
  doc = nlp(s)
  for token in doc:
    total_tokens_true+=1
    if token.pos_=='PRON':
      total_pronouns_true+=1
      if token.text.lower() in pronouns_true:
        pronouns_true[token.text.lower()]+=1
      else:
        pronouns_true[token.text.lower()]=1
    elif token.pos_=='PROPN':
      total_propernouns_true+=1
      if token.lemma_ in propernouns_true:
        propernouns_true[token.lemma_]+=1
      else:
        propernouns_true[token.lemma_]=1
```

```python
pronouns_false_sorted = collections.OrderedDict(sorted(pronouns_false.items(), key=lambda kv: kv[1], reverse=True))
propernouns_false_sorted = collections.OrderedDict(sorted(propernouns_false.items(), key=lambda kv: kv[1], reverse=True))
pronouns_true_sorted = collections.OrderedDict(sorted(pronouns_true.items(), key=lambda kv: kv[1], reverse=True))
propernouns_true_sorted = collections.OrderedDict(sorted(propernouns_true.items(), key=lambda kv: kv[1], reverse=True))
```

```python
print("for false data:") 
print(total_pronouns_false, "pronouns and", total_propernouns_false, "proper nouns out of", total_tokens_false, "tokens total")
print("pronoun frequency of", total_pronouns_false / total_tokens_false)
print("proper noun frequency of", total_propernouns_false / total_tokens_false)
print("")
print("for true data:")
print(total_pronouns_true, "pronouns and", total_propernouns_true, "proper nouns out of", total_tokens_true, "tokens total")
print("pronoun frequency of", total_pronouns_true / total_tokens_true)
print("proper noun frequency of", total_propernouns_true / total_tokens_true)
```
